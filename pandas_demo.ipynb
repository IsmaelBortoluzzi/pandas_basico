```python
country_respondants = df['Country'].value_counts()
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
python_df = pd.concat([country_respondants, country_uses_python], axis='columns', sort=False)
print(python_df.to_string())
```

```python
python_df.rename(columns={'Country': 'NumRespondants', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
```

```python
print(python_df.to_string())
```

```python
python_df['PctKnowsPython'] = (python_df['NumKnowsPython'] / python_df['NumRespondants']) * 100
print(python_df.to_string())
```

```python
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
```
